<h2>Stop Schedule</h2>

```python
schedule_id = 'a2c41bb8-21f4-4389-9ef3-e3a19a733602'
```

```python
def stop_by_schedule_id(ws, schedule_id):
    s = next(s for s in Schedule.list(ws) if s.id == schedule_id)
    s.disable()
    return s

stop_by_schedule_id(ws, schedule_id)
```
